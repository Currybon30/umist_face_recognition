```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create comparison data
clustering_results = {
   'Method': ['K-means', 'DBSCAN', 'GMM', 'Hierarchical', 'Spectral'],
   'Silhouette Score': [0.527, 0.286, 0.114, 0.457, 0.465], 
   'Number of Clusters': [24, 16, 14, 26, 24],
   'Computation Speed': [5, 3, 4, 2, 3],  # 1-5 scale (5 fastest)
   'Interpretability': [5, 4, 3, 4, 3]    # 1-5 scale (5 most interpretable)
}
```

```python
# Create DataFrame
df = pd.DataFrame(clustering_results)
```

```python
# 1. Bar plot for silhouette scores
plt.figure(figsize=(10, 6))
plt.bar(df['Method'], df['Silhouette Score'])
plt.title('Comparison of Silhouette Scores Across Clustering Methods')
plt.ylabel('Silhouette Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('silhouette_comparison.png')
plt.show()
```

```python
# 2. Dual metric comparison
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Clustering Method')
ax1.set_ylabel('Silhouette Score', color=color)
ax1.plot(df['Method'], df['Silhouette Score'], color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Number of Clusters', color=color)
ax2.plot(df['Method'], df['Number of Clusters'], color=color, marker='s')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Clustering Methods Comparison')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('dual_metric_comparison.png')
plt.show()
```

```python
# 3. Heatmap visualization
df_metrics = df.copy()
df_metrics = df_metrics.set_index('Method')
df_metrics = df_metrics.drop('Number of Clusters', axis=1)  # Remove for better scale in heatmap

plt.figure(figsize=(10, 6))
sns.heatmap(df_metrics, annot=True, cmap='YlOrRd', fmt='.3g')
plt.title('Clustering Methods Comparison Heatmap')
plt.tight_layout()
plt.savefig('clustering_comparison_heatmap.png')
plt.show()
```
